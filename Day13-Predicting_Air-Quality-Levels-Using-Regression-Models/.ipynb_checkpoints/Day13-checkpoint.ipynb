```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import cartopy.crs as ccrs
import cartopy.feature as cf
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
import xgboost as xgb
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('air_quality_ghana.csv')
df
```

```python
df.info()
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
df.describe()
```

```python
df['DateTime'] = pd.to_datetime(df['DateTime'])
```

```python
df
```

```python
df['Month'] = df['DateTime'].dt.month
df['Time'] = df['DateTime'].dt.time
df['Day'] = df['DateTime'].dt.day
```

```python
df
```

```python
X = df.drop(columns=['PM2.5', 'DateTime', 'Latitude', 'Longitude','Time','Month'])
y = df['PM2.5']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=9)
```

```python
#models 
lin_model = LinearRegression()
dec_model = DecisionTreeRegressor()
rand_model = RandomForestRegressor()
grad_model = GradientBoostingRegressor()
```

```python
# Linear model
lin_model.fit(X_train, y_train)
lin_y_pred = lin_model.predict(X_test)

print('MSE: ', mean_squared_error(y_test, lin_y_pred))
print('RMSE: ', np.square(mean_squared_error(y_test, lin_y_pred)))
print('MAE: ', mean_absolute_error(y_test, lin_y_pred))
```

```python
# Decision Tree model
dec_model.fit(X_train, y_train)
dec_y_pred = dec_model.predict(X_test)

print('MSE: ', mean_squared_error(y_test, dec_y_pred))
print('RMSE: ', np.square(mean_squared_error(y_test, dec_y_pred)))
print('MAE: ', mean_absolute_error(y_test, dec_y_pred))
```

```python
# Random Forest model
rand_model.fit(X_train, y_train)
rand_y_pred = rand_model.predict(X_test)

print('MSE: ', mean_squared_error(y_test, rand_y_pred))
print('RMSE: ', np.square(mean_squared_error(y_test, rand_y_pred)))
print('MAE: ', mean_absolute_error(y_test, rand_y_pred))
```

```python
# Gradient Boosting model
grad_model.fit(X_train, y_train)
grad_y_pred = grad_model.predict(X_test)

print('MSE: ', mean_squared_error(y_test, grad_y_pred))
print('RMSE: ', np.square(mean_squared_error(y_test, grad_y_pred)))
print('MAE: ', mean_absolute_error(y_test, grad_y_pred))
```

```python
# Gradient Boosting model
xgb_model = xgb.XGBRegressor(objective='reg:squarederror', eval_metric='rmse')
xgb_model.fit(X_train, y_train)
xgb_y_pred = xgb_model.predict(X_test)

print('MSE: ', mean_squared_error(y_test, xgb_y_pred))
print('RMSE: ', np.square(mean_squared_error(y_test, xgb_y_pred)))
print('MAE: ', mean_absolute_error(y_test, xgb_y_pred))
```
